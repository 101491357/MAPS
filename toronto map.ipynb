```python
import pandas as pd
import folium

def init_map():
    # Read the CSV file into a DataFrame
    TorontoPolice_df = pd.read_csv(r"C:\Users\nanaa\Downloads\KSI (1) (1).csv")

    # Create a Folium map object
    my_map = folium.Map(location=[43.77092102, -79.30225887], zoom_start=13)

    # Add markers for each accident location
    for index, row in TorontoPolice_df.iterrows():
        # Extract latitude, longitude, and accident location
        latitude = row['LATITUDE']
        longitude = row['LONGITUDE']
        accloc = str(row['ACCLOC'])  # Convert to string to handle float values
        
        # Define marker color based on whether AccLoc represents an intersection
        if 'At Intersection' in accloc:
            marker_color = 'black'  # If AccLoc contains 'Intersection', set marker color to black
        else:
            marker_color = 'brown'  # Otherwise, set marker color to brown
        
        # Add a marker for each accident location with the specified color
        folium.Marker(location=[latitude, longitude], popup=accloc, icon=folium.Icon(color=marker_color)).add_to(my_map)

    return my_map

if __name__ == "__main__":
    my_map = init_map()
    my_map.save("toronto_accidents_map.html")  # Save the map to an HTML file
```
